```python
def avg(a):
    return np.mean(a, axis=0)

def naive(a,t):
    return a[a.shape[0]-t,:]

def convert_date(t):
    d = datetime.strptime(t, '%Y-%m-%d %H:%M:%S')
    day = int(d.strftime('%d'))
    hour = int(d.strftime('%H'))
    minute = int(d.strftime('%M'))
    index = (day-1)*144 + ((minute/10)+1) + hour*6
    return index

def convert_time(t):
    day = t / 144
    minutes = (t - day*144)*10
    hour = minutes / 60
    minute = minutes - hour*60
    return (day,hour,minute)
```

```python
frames = []
mypath = '/media/sahabi/Shared Storage/season_1/training_data/weather_data/'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
for i,files in enumerate(onlyfiles):
    data = mypath+files
    order_data = pd.read_csv(data,sep='\t',names=['Time','Weather','temperature','PM2.5'])
    frames.append(order_data)
    if i ==140:
        break

result = pd.concat(frames)
order_data=result.fillna(0)
order_data['t_ind'] = order_data.Time.apply(convert_date)
```
